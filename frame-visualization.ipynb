```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

### Using plot (axes of the coordinate frames are lines)

```python
%matplotlib notebook
fig = plt.figure()
ax = plt.axes(projection="3d")

twozeros = np.zeros(2)
ax.plot3D([0, 1], twozeros, twozeros)
ax.plot3D(twozeros, [0, 1], twozeros)
ax.plot3D(twozeros, twozeros, [0, 1])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

### Using quiver (axes of the coordinate frames are arrows)

```python
%matplotlib notebook
fig = plt.figure()
ax = plt.axes(projection="3d")

i = [1,0,0]
j = [0,1,0]
k = [0,0,1]
ax.quiver(0,0,0,i[0],i[1],i[2])
ax.quiver(0,0,0,j[0],j[1],j[2])
ax.quiver(0,0,0,k[0],k[1],k[2])

ax.set_xlabel('N')
ax.set_ylabel('E')
ax.set_zlabel('D')

ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)

ax.invert_yaxis()
ax.invert_zaxis()

ax.view_init(20,45)

plt.show()
```
